The value for the Rydberg constant, $\mathcal{R}$, can be determined by measuring the wavelength of several lines in the spectrum of light from a hydrogen spectrul tube. The spectral lines that belong to the Balmer series will obery the formula:
\begin{equation}
\frac{1}{\lambda} = \mathcal{R}\left(\frac{1}{n^2} - \frac{1}{m^2}\right),
\end{equation}
where $n=2$ and $m=3,4,5,\ldots$.

Calibration using a HeNe laser:

```python
# Importing data and converting it into a pandas table
calibration_data = pd.read_csv('Balmer Lab Data/calibration5.csv', skiprows=14, names=['wavelength', 'count'])
xdata = np.array(calibration_data['wavelength'])
ydata = np.array(calibration_data['count'])

# Plotting the counts as a function of wavelength
fig, ax = plt.subplots()
ax.scatter(xdata, ydata, marker='x', color='black', s=20)
plt.show()
```
